The topic of classifier in today's lab, SVMs make for really good linear separators. Let us look at an example which has linearly separable data:

```python
np.random.seed(0)

ar = np.vstack(     [\
                    np.random.multivariate_normal(np.array([1, 1]), 1.5 * np.array([[2, -1], [-1, 2.0]]), size = 50, ),\
                    np.random.multivariate_normal(np.array([3, 3]), 2 * np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 50, )
                    ]\
              )

testAr = np.vstack(   [\
                      np.random.multivariate_normal(np.array([1, 1]), np.array([[0.5, -0.25], [-0.25, 0.5]]), size = 500, ),\
                      np.random.multivariate_normal(np.array([3, 3]), np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 500, )
                      ]\
                  )
testy = np.array([0] * int((testAr.shape[0]/2)) + [1] * int((testAr.shape[0]/2)))

X = ar
y = np.array([0] * int((ar.shape[0]/2)) + [1] * int((ar.shape[0]/2)))

def plotDecisionBoundary(X, y, pair, clf):
  x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
  y_min, y_max = X[:, pair[1]].min() - 1, X[:, pair[1]].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                      np.arange(y_min, y_max, 0.1))

  y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
  y_pred = y_pred.reshape(xx.shape)
  plt.figure(figsize=(5,4))
  plt.contourf(xx, yy, y_pred, alpha=0.4)
  plt.scatter(X[:, pair[0]], X[:, pair[1]], c = y, s = 50, edgecolor='k')
  # plt.legend()
  plt.gcf().set_dpi(130)
  plt.show()

def boundaryExp() :
  clf = svm.LinearSVC()
  pair = [0, 1]
  clf.fit(X[:, pair], y)
  plotDecisionBoundary(X, y, pair, clf)

  plt.show()

boundaryExp()
```

## Intuitive introduction to SVMs:

![SVM](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1526288453/index3_souoaz.png)

SVMs are an iterative approach of trying to find the optimal hyperplane that divides the multidimentional space into different classes present in the dataset.


**Hyperplanes:** These are the decision planes that separate the objects of classes that we are trying to classify.

**Support Vectors** : Support vectors are the points from dataset that are closest to the hyperplane that divides the dataset.

**Margin**:  The gap between the closest support vectors from the different class along the direction perpendicular to the hyperplane. Simply put, it is the sum of perpendicular distance of the support vector of each class to the hyperplane.

Let us try to fall back to the well known dataset and see how SVMs perform on the iris dataset

```python
iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
irisData = pandas.DataFrame(\
    data = np.hstack((X,y.reshape(y.shape[0], 1), [[iris["target_names"][int(classIdx)]] for classIdx in y])), \
    columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', "Class", "ClassName"])
irisData.sample(n = 10, random_state = 1)
```

# The Kernel Trick

The true potential of SVMs is unleashed when they are combined with kernels.

## Kernels : An intuitive explanation

Kernel methods are essentially counting on using the training data (say $i^{th}$ example $(x_i, y_i)$ ) itself in a more straightforward way and learning a corresponding weight ($w_i$) for that example. Rather than trying to learn a fixed set of parameters which is done typically.
Depending on the kind of kernel used, we can virtually project the training data in a higher dimension to make it easier for the classifier to classify them.


```python
from sklearn.datasets import make_circles
X, y = make_circles(200, factor=.1, noise=.1)

clf = svm.SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='tab20')
plt.gcf().set_dpi(110)
```

```python
plt.style.use("default")
clf = svm.LinearSVC()
pair = [0, 1]
clf.fit(X[:, pair], y)
plotDecisionBoundary(X, y, [0, 1], clf)
```

However, if we artificially add another dimention to the dataset of the form:

$z = x^2 + y^2$
we can clearly see a hyperplane that can distinguish both the datasets:

```python
Z = np.array([[i[0]**2 + i[1]**2] for i in X])
```

```python
X_new = np.hstack((X, Z))
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data = [go.Scatter3d(
    x = X_new[:, 0],
    y = X_new[:, 1],
    z = X_new[:, 2],
    mode = "markers",
    marker = {
        "color" : y,
        "line": {"width" : 4, "color":'DarkSlateGrey'},
        "colorscale": "viridis"},
)])

fig.show()
```

That simple trick has helped us to get another dimension in which the data is linearly separable by a hyperplane (in this case, a 2d plane)

Certainly, using a kernel has increased our accuracy on the iris dataset.

**Answer 1:-**

Drawing the decision boundary, margins, and identifying support vectors:

- Decision Boundary: The decision boundary is the line that best separates the two classes (black diamonds and white diamonds) with the maximum margin. In this case, since we're considering a linear hard margin SVM, the decision boundary will be a straight line. To determine the exact equation of the line, we would need to solve the SVM optimization problem, which aims to find the hyperplane that maximizes the margin while correctly classifying all training samples.
- Margins: The margins are the parallel lines on either side of the decision boundary that define the region within which no training samples are allowed. The distance between the margins is called the margin width, and the SVM aims to maximize this width. The margins are defined by the support vectors.
- Support Vectors: Support vectors are the training samples that lie on the margin or violate the margin constraint. They are the critical data points that determine the position and orientation of the decision boundary. Other training samples that are further away from the margin do not influence the decision boundary. In the given scenario, we need to identify the black and white diamonds that lie closest to the potential decision boundary.

**Answer 2:-**

Training error rate:

The training error rate is the proportion of training samples that are misclassified by the SVM model. Since we are using a hard margin SVM, it aims to achieve zero training error, meaning it tries to classify all training samples correctly. If the data is linearly separable, the hard margin SVM will find a decision boundary that perfectly separates the classes, resulting in a training error rate of 0.

**Answer 3:-**

Sample removal affecting the decision boundary:

The removal of a support vector is most likely to change the decision boundary. This is because support vectors directly define the position and orientation of the decision boundary. Removing other training samples that are not support vectors will typically not affect the decision boundary, as they are not involved in its definition.

**Answer 4:-**


Leave-one-out error rate:

The leave-one-out error rate is a cross-validation technique where we repeatedly train the SVM model on all but one training sample and then test it on the left-out sample. This process is repeated for each training sample, and the leave-one-out error rate is the proportion of times the model misclassifies the left-out sample. In the case of a hard margin SVM, the leave-one-out error rate is often related to the number of support vectors.

Specifically, for a linearly separable dataset, the leave-one-out error rate is bounded by the number of support vectors divided by the number of training examples. This is because if we remove a training sample that is not a support vector, the decision boundary will not change, and the model will still classify the left-out sample correctly. However, if we remove a support vector, the decision boundary might change, leading to a misclassification of the left-out sample.
